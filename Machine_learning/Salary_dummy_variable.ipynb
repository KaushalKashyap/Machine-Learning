```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
```

```python
df=pd.read_excel('Salary_dummy_raw_data.xlsx')
df
```

```python
#Should convert the x(Education, Dtype=int64) Dtype in string. 
df['Education']=df['Education'].apply(str)
```

```python
df.info()
```

```python
#create dummy variable.(purpose is to differentiate different type of data in one colum.ex- Martial status= m,d,s,) Now using Dummy variable it will create four dummy column like M, D,S.
df_dummies=pd.get_dummies(df['Education'],drop_first=True)
#In Ml it is called one-hot-encoding, Here the drop_first will check if condition is true then 1 otherwise it will move out
```

```python
df_dummies
```

```python
#if you want to drop any row from dummies
#df_dummies.drop('4',axis=1,inplace=True)
```

```python
#Now after creating dummy variable , want to check how its looks like
df=pd.concat([df_dummies,df['Salary']],axis=1)
```

```python
df.head()
```

```python
#Now come to run for Regression analysis
X=sm.add_constant(df[['2','3','4']])
y=df['Salary']
```

```python
model=sm.OLS(y,X).fit()
```

```python
model.summary2()
```

```python
model.params
```
