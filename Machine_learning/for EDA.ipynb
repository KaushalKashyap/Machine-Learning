```python
# although we already know about quartiles, mean, median, mode, min, max in 'describe function'.
# This is for manual computation But there is function for it get the value
# the output value is only for tip column.
print(df.Per_hour_Electricity.mean())
print(df.Per_hour_Electricity.median())
print(df.Per_hour_Electricity.mode())
print(df.Per_hour_Electricity.quantile(.25))
print(df.Per_hour_Electricity.quantile(.50))
print(df.Per_hour_Electricity.quantile(.75))
print(df.Per_hour_Electricity.max())
print(df.Per_hour_Electricity.min())
```

```python
from scipy.stats import kurtosis,skew
print(skew(df.Per_hour_Electricity,bias=False))
print(kurtosis(df.Per_hour_Electricity,bias=False))
```
