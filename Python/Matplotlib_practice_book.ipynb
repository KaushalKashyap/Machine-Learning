```python
import matplotlib.pyplot as plt
```

```python
#these are demo of data. which 
week_day=[1,2,3,4,5,6,7]
temp=[50,51,52,48,47,49,46] 
```

```python
#here we have used plt .plot with parameter Week_day and Temp. where the two data is stored.
plt.plot(week_day, temp)
```

```python
#here we are using xlabel and ylabel (with parameter) will show what values are in x axis and y axis.
plt.plot(week_day, temp)
plt.xlabel("Days")
plt.ylabel("Temp")
```

```python
# here in plt.plot(arg1,arg2,....). inside that we can check how many argument we can pass by using shortcut key(shift+tab)
plt.plot(week_day, temp, color="r")
plt.xlabel("Days")
plt.ylabel("Temp")
#here we can also put a title for the graph 
plt.title("Weather Information")
```

```python
#here we can change the linestyle over here by using kwargs linestyle="--"
plt.plot(week_day, temp, color = "r" , linestyle = "--")
plt.xlabel("Days")
plt.ylabel("Temp")
plt.title("Weather Information")
```

```python
plt.plot(week_day, temp, color = "r" , linestyle = "--") #fmt = '[marker][line][color]'
plt.xlabel("Days")
plt.ylabel("Temp")
plt.title("Weather Information")
```

```python
# The marker kwargs will shows that the points where stated on graph.

plt.plot(week_day , temp , color = 'g' , linestyle = '--' , marker = 'o')
plt.xlabel('Day')
plt.ylabel('temp')
plt.title('weather information')
```

```python
# Format string 
#'[marker][line][color]'
#here we are kwargs in single string attached.for marker o for line -- for linecolor b. so it will be o--b

plt.plot(week_day , temp , "o--b")
plt.xlabel('Day')
plt.ylabel('temp')
plt.title('weather information')
```

```python
# no need to follow the sequence of [marker][line][color]'
plt.plot(week_day , temp , "--<g")
plt.xlabel('Day')
plt.ylabel('temp')
plt.title('weather information')
```

```python
# to increase marker size in graph use markersize 
plt.plot(week_day , temp , "--og" , markersize = 10)
plt.xlabel('Day')
plt.ylabel('temp')
plt.title('weather information')
```

```python
#if we have more than two kind of data. then in that case we will plot three time.
days = [1,2,3,4,5,6,7]
max_t = [50,51,52,53,54,57,70]
min_t = [45 , 42,44,46,43,41,40]
avg_t = [45,48, 48, 46,40,42,41]
```

```python
plt.plot(days , max_t)
plt.plot(days , min_t)
plt.plot(days , avg_t)
```

```python
plt.plot(days , max_t)
plt.plot(days , min_t)
plt.plot(days , avg_t)
plt.xlabel("Days")
plt.ylabel("Temps")
plt.title("Temp VS Days")
```

```python
#now here if want to put markersize , marker . so we have to go through every different data. 
#like for line days, max_t, "--oc"
#like for line days, min_t, "--om"
##like for line days, max_t, "--oy"
plt.plot(days , max_t)
plt.plot(days , min_t)
plt.plot(days , avg_t)
plt.xlabel("Days")
plt.ylabel("Temps")
plt.title("Temp VS Days")

plt.plot(days , max_t , "--oc" , markersize = 10)
plt.plot(days , min_t , "--om" , markersize = 10)
plt.plot(days , avg_t , "--oy" , markersize = 10)
```

```python
#here in previous the data is very dummy, means rear to find which line denotes which data. so for that legend comes in light.
#legend will bracket a new corner part of graph where it shows which color defines which data.
plt.plot(days , max_t , label = "Maximum")
plt.plot(days , min_t , label = "Minimum")
plt.plot(days , avg_t , label = "Average")
plt.xlabel("Days")
plt.ylabel("Temps")
plt.title("Temp VS Days")
plt.legend()
```

```python
#if you forgot or want to know something. then go for help.
help(plt.plot)
```

```python
# here we use grid fn for background page of the graph.
plt.plot(days , max_t , label = "Maximum")
plt.plot(days , min_t , label = "Minimum")
plt.plot(days , avg_t , label = "Average")
plt.xlabel("Days")
plt.ylabel("Temps")
plt.title("Temp VS Days")
plt.legend()
plt.grid()
```

```python
# if you want to move legend small bracket to some other corner. then pass the parameter in legend(loc="lower left")
plt.plot(days , max_t , label = "Maximum")
plt.plot(days , min_t , label = "Minimum")
plt.plot(days , avg_t , label = "Average")
plt.xlabel("Days")
plt.ylabel("Temps")
plt.title("Temp VS Days")
plt.legend( loc = "upper left")
plt.grid()
```

```python
blood_sugar = [113,85,90,150,149,88,93,115,135,80,77,82,129]
```

```python
#here bin will simply draw 4 bar histogramatically and those bar will define the bin value which is 77,100,125,150.
#reason behind using bin is histogram is very dummy, very difficult to find difference if data was close. with the help of  bin
# we can define 4 or five interval using data range. and with data range it will show the below comment derived data.
plt.hist(blood_sugar, bins = [77, 100 , 125 , 150])
#no diabetics = 80 - 100
#pre diabetic = 100 - 125
#have diabetic = 125 - 150
```

```python
#if you want to plot a histogram with different color. you can do it by by giving them two different color 
blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]

plt.hist([blood_sugar_men, blood_sugar_women] , bins = [77, 100 , 125 , 150] , 
         color = ["g" , "r"], label = ["mens" , "women"])
```

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
```

```python
plt.pie(exp_vals , labels = exp_labels)
plt.show()
```

```python
# Percentage
# Exact values
# Size
```

```python
#here you can give radius  value in parameter.
plt.pie(exp_vals , labels = exp_labels , radius = 2)
plt.show()
```

```python
#if you want to show percentage inside the pie chart, below are the syntax
plt.pie(exp_vals , labels = exp_labels , radius = 2 , autopct = '%0.2f%%')
plt.show()
```

```python
#here explode will actually differentiate every part of data
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]


plt.pie(exp_vals , labels = exp_labels , radius = 2 , autopct = '%0.0f%%' , explode = [0.5,0,0.3,0,0.1])
plt.show()
plt.savefig("pie.pdf")
```

```python
plt.pie(exp_vals , labels = exp_labels , radius = 2 , autopct = '%0.0f%%' , explode = [0.5,0,0.3,0,0.1])
plt.show()
plt.savefig("pie.png")
```
