```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
```

```python
df = pd.read_excel("Salary_dummy_raw_data.xlsx", index_col = None)
df.info()
```

```python
# df.describe()
df.head(20)
```

```python
df['Education'] = df['Education'].apply(str)
```

```python
df.info()
```

```python
# create dummy variables

df_dummies = pd.get_dummies(df['Education'], drop_first=True)
# drop_first has to be explicitly given. 

# in ML this is called one-hot encoding (sklearn has one-hot encoding)
```

```python
df_dummies
```

```python
# df_dummies.drop('4', axis=1, inplace=True)
```

```python
df = pd.concat([df_dummies, df['Salary']], axis = 1) 
# just for presentation
```

```python
df.head()
```

```python
X = sm.add_constant(df[['2', '3', '4']])
y = df['Salary']
```

```python
model = sm.OLS(y, X).fit()
```

```python
model.summary()
```

```python
model.params
```
